### 4.Convert the data into a unified format, such as a common dataframe or data structure, to enable seamless analysis.

```python
df.drop(columns='ADDRESSLINE2',inplace=True)
```

```python
df['STATE'].fillna("Unknown", inplace=True)
df['TERRITORY'].fillna("Unknown", inplace=True)
df['POSTALCODE'].fillna("00000", inplace=True)
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df_xl.duplicated().sum()
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], errors='coerce')
df[['PRODUCTLINE','DEALSIZE','STATUS']] = df[['PRODUCTLINE','DEALSIZE','STATUS']].astype('category')
```

```python
df.dtypes
```

```python
df_xl.dtypes
```

```python
months = ['JANUARY', 'FEBRUARY', 'MARCH', 'APRIL', 'MAY', 'JUNE', 'JULY','AUGUST', 'SEPTEMBER', 'OCTOBER', 'NOVEMBER', 'DECEMBER']
for month in months:
    df_xl[month] = pd.to_numeric(df_xl[month], errors='coerce')
```

```python
df_xl.dtypes
```

```python
df_xl.describe()
```

```python
print(df_xl.isnull().sum())
```

```python
df_xl.fillna(0, inplace=True)
print(df_xl.isnull().sum())
```

### 5.Perform data transformation tasks, such as merging multiple datasets,splitting columns, or deriving new variables.

```python
df_xl['TOTAL_SALES'] = df_xl[['JANUARY', 'FEBRUARY', 'MARCH', 'APRIL', 'MAY', 'JUNE','JULY', 'AUGUST', 'SEPTEMBER', 'OCTOBER', 'NOVEMBER', 'DECEMBER']].sum(axis=1)
```

```python
df['TOTAL_ORDER_VALUE'] = df['QUANTITYORDERED'] * df['PRICEEACH']
```

```python
df['ORDER_YEAR'] = df['ORDERDATE'].dt.year
df['ORDER_MONTH'] = df['ORDERDATE'].dt.month
```

### 6.Analyze the sales data by performing descriptive statistics, aggregating data by specific variables, or calculating metrics such as total sales, average order value, orproduct category distribution.

```python
total_sales = df['SALES'].sum()
print(f"Total Sales: {total_sales}")
```

```python
average_order_value = df['SALES'].mean()
print(f"Average Order Value: {average_order_value}")
```

```python
sales_by_product_line = df.groupby('PRODUCTLINE')['SALES'].sum()
print(f"Sales by Product Line:\n {sales_by_product_line}")
```

```python
sales_by_territory = df.groupby('TERRITORY')['SALES'].sum()
print("Sales by Territory:")
print(sales_by_territory)
```

```python
sales_by_year = df.groupby('ORDER_YEAR')['SALES'].sum()
print("Sales by Year:")
print(sales_by_year)
```

```python
sales_by_month = df.groupby('ORDER_MONTH')['SALES'].sum()
print("Sales by Month:")
print(sales_by_month)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# sales_by_product_line = df.groupby('PRODUCTLINE')['TOTAL_ORDER_VALUE'].sum()
sales_by_product_line.plot(kind='bar')
plt.title('Total Sales by Product Line')
plt.xlabel('Product Line')
plt.ylabel('Total Sales')
plt.show()
```

```python
# Pie chart for sales status distribution
sales_status = df['DEALSIZE'].value_counts()
sales_status.plot(kind='pie', autopct='%1.1f%%')
plt.title('Sales Status Distribution')
plt.ylabel('')
plt.show()
```

```python
monthly_sales = df.groupby('ORDER_MONTH')['TOTAL_ORDER_VALUE'].sum()
monthly_sales.plot(kind='bar')
plt.title('Total Sales by Month')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
sales_by_territory = df_xl.groupby('SALES_BY_REGION')['TOTAL_SALES'].sum()
sales_by_territory.plot(kind='bar')
plt.title('Total Sales by Territory')
plt.xlabel('Territory')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
